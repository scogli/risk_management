```python
import pandas as pd 
import numpy as np
from scipy.stats import norm
from tqdm import tqdm
```

```python
portfolio = pd.read_excel("portfolio.xlsx",sheet_name = "portfolio",index_col = "Date")
```

```python
parameters = pd.read_excel("portfolio.xlsx",sheet_name = "garch_fit",index_col = "parameters")
```

```python
n_iterations = 1000000
t_intervals = 20
```

```python
days_list = []
for i in range(1,21):
    days_list.append("day_" + str(i))
```

```python
days_list
```

```python
variance = pd.DataFrame(np.linspace(portfolio["total_garch_variance"].iloc[1],portfolio["total_garch_variance"].iloc[1],num = n_iterations),columns= ["day_0"])
```

```python
variance["day_1"] = np.linspace(portfolio["total_garch_variance"].iloc[0],portfolio["total_garch_variance"].iloc[0],num = n_iterations)
```

```python
inovation = pd.DataFrame(norm.ppf(np.random.rand(n_iterations,t_intervals),portfolio["mapped_portfolio_returns"].mean(),portfolio["mapped_portfolio_returns"].std()),columns = days_list)
```

```python
inovation[days_list[0]] = inovation[inovation.columns[0]] * np.sqrt(variance["day_1"])
```

```python
beta1_tot = parameters["total_fit"].loc["beta1"]
beta2_tot = parameters["total_fit"].loc["beta2"]
omega_tot = parameters["total_fit"].loc["omega"]
alpha_tot = parameters["total_fit"].loc["alpha"]
gamma_tot = parameters["total_fit"].loc["gamma"]
miu_tot = parameters["total_fit"].loc["miu"]
rho1_tot = parameters["total_fit"].loc["rho1"]
rho2_tot = parameters["total_fit"].loc["rho2"]
rho3_tot = parameters["total_fit"].loc["rho3"]
```

```python
identity_function = inovation[inovation < 0].fillna(0)
```

```python
identity_function = identity_function **2 
```

```python
for i in range(2,21):
    variance["day_" + str(i)] = omega_tot + beta1_tot * variance["day_"+str(i-1)] + beta2_tot * variance["day_"+str(i-2)] + alpha_tot * inovation["day_" + str(i-1)] + gamma_tot * identity_function["day_"+str(i-1)]
```

```python
variance
```

```python
returns = pd.DataFrame(np.linspace(portfolio["mapped_portfolio_returns"].iloc[2],portfolio["mapped_portfolio_returns"].iloc[2], num = n_iterations),columns = ["day_-2"])
```

```python
returns["day_-1"] = np.linspace(portfolio["mapped_portfolio_returns"].iloc[1],portfolio["mapped_portfolio_returns"].iloc[1], num = n_iterations)
```

```python
returns["day_0"] =  np.linspace(portfolio["mapped_portfolio_returns"].iloc[1],portfolio["mapped_portfolio_returns"].iloc[1], num = n_iterations)
```

```python
for i in range(1,21):
    returns["day_"+str(i)] = miu_tot + rho1_tot * returns["day_"+str(i-1)] +  rho2_tot * returns["day_"+str(i-2)] + rho3_tot * returns["day_"+str(i-3)] + inovation["day_"+str(i)]
```

```python
returns
```

```python
interm_step = returns[days_list] + 1 
```

```python
return_20d = []
for i in tqdm(interm_step.index):
    return_20d.append(np.prod(interm_step.loc[i])-1)
```

```python
returns["20d_returns"] = return_20d
```

```python
alpha = 0.01
```

```python
np.quantile(returns["20d_returns"], 1- alpha)
```
