```python
returns_usd = returns[['srt_ROST', 'srt_AAPL', 'srt_MA', 'srt_ANSS', 'srt_TSCO', 'srt_NFLX','srt_AMD','srt_GE', 'srt_F','srt_^GSPC']]
```

```python
returns_gbp = returns[['srt_EZJ.L', 'srt_SDR.L','srt_BATS.L','srt_VOD.L','srt_BA.L','srt_^FTSE']]
```

```python
returns_eur = returns[['srt_ADS.DE', 'srt_DPW.DE', 'srt_BAS.DE','srt_RI.PA', 'srt_BNP.PA', 'srt_AI.PA','srt_^GDAXI','srt_^FCHI']]
```

```python
betas_usd = pd.DataFrame(returns_usd.cov()["srt_^GSPC"]/returns_usd["srt_^GSPC"].var())
```

```python
betas_usd.columns = ["Beta"]
```

```python
betas_usd = betas_usd.T
```

```python
betas_usd
```

```python
betas_gbp = pd.DataFrame(returns_gbp.cov()["srt_^FTSE"]/returns_gbp["srt_^FTSE"].var())
```

```python
betas_gbp.columns = ["Beta"]
```

```python
beta_gbp = betas_gbp.T
```

```python
beta_gbp
```

```python
residuals_usd = pd.DataFrame()

for columns in returns_usd[returns_usd.columns[0:8]]:
    residuals_usd[columns] = returns_usd[columns] - returns_usd["srt_^GSPC"] * betas_usd[columns].values
```

```python
residuals_gbp = pd.DataFrame()

for columns in returns_gbp.columns[0:5]:
    residuals_gbp[columns] = returns_gbp[columns] - returns_gbp["srt_^FTSE"] * beta_gbp[columns].values
```
